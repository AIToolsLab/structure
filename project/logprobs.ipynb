```python
import os
import openai

openai.Model.list()
```

```python
question_w_outline = '''
Write a short essay given this outline:
• Caught travel bug at age 5 when grandparents brought to France 
• Visited 29 countries
• Experienced the Eiffel Tower in Paris at age 5
• Fed pigeons in Piazza San Marco at age 8
• Saw Stonehenge at age 13
• Walked the Great Wall of China at age 13

'''
```

```python
sentences = [
    'When I was very little, I caught the travel bug.',
    'It started after my grandparents first brought me to their home in France and I have now been to twenty-nine different countries.',
    'Each has given me a unique learning experience.',
    'At five, I marveled at the Eiffel Tower in the City of Lights.',
    'When I was eight, I stood in the heart of Piazza San Marco feeding hordes of pigeons, then glided down Venetian waterways on sleek gondolas.',
    'At thirteen, I saw the ancient, megalithic structure of Stonehenge and walked along the Great Wall of China, amazed that the thousand-year-old stones were still in place.'
]
```

```python
def all_unique_pos(lst, element):
    res = []
    for i in range(len(lst)):
        new_list = lst[:]
        new_list.pop(lst.index(element))
        res.append(new_list[:i] + [element] + new_list[i:])
    return res
```

```python
combinations = all_unique_pos(sentences, 'When I was very little, I caught the travel bug.')
```

```python
original_text = ' '.join(sentences)
original_text
```

```python
full_request = question_w_outline + original_text
full_request
```

```python
def full_request_template(question, sents):
    return question + ' '.join(sents)
```

```python
request = full_request_template(question_w_outline, sentences)
request
```

```python
import joblib
from joblib import Memory
```

```python
memory = Memory("./joblib_cache", verbose = 0)

@memory.cache
def get_response(full_request):
  response = openai.Completion.create(
      model="text-davinci-003",
      prompt = full_request,
      temperature=0.7,
      max_tokens=256,
      top_p=1,
      frequency_penalty=0,
      presence_penalty=0,
      logprobs=10,
      echo=True
    )
  return response
```

```python
def get_all_responses(options):
    response = []
    for x in options:
        response.append(get_response(full_request_template(question_w_outline, x)))
    return response
```

```python
result = get_all_responses(all_unique_pos(sentences, 'When I was very little, I caught the travel bug.'))
result
```

Compute total logprobs of the original text

```python
def compute_log_probs(question_w_outline, original_text, response):
    # 1. Start point: text offset of the first word from our original essay
    start_point = len(question_w_outline)
    # 2. Length of the original text
    start_index = response.choices[0].logprobs.text_offset.index(start_point)
    # 3. End point = Start point + Length of the original text
    len_original_text = len(original_text)
    # 4. Compute total logprobs of the original text
    end_point = start_point + len_original_text
    # 5. end index
    end_index = response.choices[0].logprobs.text_offset.index(end_point)
    # total
    total = 0
    for x in range(start_index, end_index + 1):
        total = total + response.choices[0].logprobs.token_logprobs[x]
    return total


    
```
